```python
url_gdp="https://en.wikipedia.org/wiki/List_of_Indian_states_and_union_territories_by_GDP"
web=wd.Chrome()
web.get(url_gdp)
content=bs(web.page_source,'html.parser')
tables = content.find_all('table', class_='wikitable')
gdp_22_23=[]
table_heading_22_23=[]
table_index=0
if len(tables)>table_index:
    gdp_22_23_table=tables[table_index]
else:
    print("can't fine the specific table mentioned")
for i in gdp_22_23_table.find_all('th'):
    table_heading_22_23.append(i.text.strip())
for i in gdp_22_23_table.find_all('td'):
    cleaned_data=i.text.replace('\u202f*','').strip()
    gdp_22_23.append(cleaned_data)
gdp_22_23
gdp_22_23_data={}
for i in range(len(table_heading_22_23)):
    gdp_22_23_data[table_heading_22_23[i]]=gdp_22_23[i::3]
gdp_22_23_data
gdp_22_23=pd.DataFrame(gdp_22_23_data)
print(gdp_22_23.head())
web.quit()
```

```python
url_gdp="https://en.wikipedia.org/wiki/List_of_Indian_states_and_union_territories_by_GDP"
web=wd.Chrome()
web.get(url_gdp)
content=bs(web.page_source,'html.parser')
tables = content.find_all('table', class_='wikitable')
table_data=[]
table_heading=[]
table={}
table_index=2
if len(tables) > table_index:
    specific_table = tables[table_index]
else:
    print(f"There are only {len(tables)} tables on the page.")
for i in specific_table.find_all('th'):
    table_heading.append(i.text.strip())
for i in specific_table.find_all('td'):
    table_data.append(i.text.strip())
statename=[]
for i in table_data:
    cleaneddata=i.replace('\u202f*','')
    statename.append(cleaneddata)
years=table_heading[0:13]
for i in range(len(years)):
    table[years[i]]=statename[i::13]
gdp_decade=pd.DataFrame(table)
print(gdp_decade.head())
web.quit()
```

```python
url_litracy_rate="https://en.wikipedia.org/wiki/List_of_Indian_states_and_union_territories_by_literacy_rate"
web=wd.Chrome()
web.get(url_litracy_rate)
content=bs(web.page_source,'html.parser')
tables=content.find_all('table', class_='wikitable')
litracy_data=[]
litracy_heading=[]
table_index=0
litracy_data_dict={}
specific_table=tables[table_index]
for i in specific_table.find_all('td'):
    litracy_data.append(i.text.strip())
litracy_data=litracy_data[7::]
for i in specific_table.find_all('th'):
    litracy_heading.append(i.text.strip())
litracy_heading.remove('Census 2011[2]')
litracy_heading.remove('NSO survey (2017)[3]')
litracy_heading[4]='N_AVERAGE'
litracy_heading[5]='M_AVERAGE'
litracy_heading[6]='F_AVERAGE'
litracy_data.insert(219,'None')
litracy_data.insert(220,'None')
for i in range(len(litracy_heading)):
    litracy_data_dict[litracy_heading[i]]=litracy_data[i::7]
litracy_rate=pd.DataFrame(litracy_data_dict)
print(litracy_rate.head())
web.quit()
```

```python
url_unemployement_rate="https://en.wikipedia.org/wiki/List_of_states_and_union_territories_of_India_by_unemployment_rate"
web=wd.Chrome()
web.get(url_unemployement_rate)
content=bs(web.page_source,'html.parser')
tables = content.find_all('table', class_='wikitable')
unemployement_data=[]
unemployement_heding=[]
unemployement_data_dict={}
table_index=0
if len(tables) > table_index:
    specific_table = tables[table_index]
else:
    print('table index not found')
for i in specific_table.find_all('th'):
    unemployement_heding.append(i.text.strip())
for i in specific_table.find_all('td'):
    unemployement_data.append(i.text.strip())
for i in range(len(unemployement_heding)):
    unemployement_data_dict[unemployement_heding[i]]=unemployement_data[i::2]
unemployement_rate=pd.DataFrame(unemployement_data_dict)
print(unemployement_rate.head())
web.quit()
```
